def media(arreglo):

- La mediana: la funcion de mediana crea un arreglo con la funcion sorted que lo organiza de forma ascendente con los datos del arreglo creado anteriormente y luego se saca la longitud del arreglo, de ahi creamos un condicional que dividirá la longitud en 2 y si el resto es distinto a 0, su longitud va ser impar, entonces se crea un indice y calculara el indice del elemento central y devuelve el valor de la posicion que marcaba el indice si al dividir es par va a calcular el indice superior del par central va tomar los dos valores centrales, los suma y divide la suma entre 2 y eso produce la mediana si es par

```
def mediana(arreglo):
  arreglo_ordenado = sorted(arreglo);
  len_arreglo = len(arreglo);

  if(len_arreglo % 2 != 0):

- La moda: en esta funcion lo que hace es crear una variable con valores unicos entonces se usa la funcion set que permite eliminar los valores repetidos y de ahi se crea una variable base que se usará para guardar la mayor cantidad de apariciones encontrada y comienza en 0 porque al principio no se ha contado nada de ahi inicia un bucle que recorre cada valor único en el conjunto de ahi usa el método count de las listas para contar cuántas veces aparece valor en el arreglo original ya luego compara cuántas veces aparece este valor con la mayor cantidad encontrada hasta el momento si la condicion cumple actualizara la moda para guardar el valor que mas se repite hasta ahora y actualiza la base para guardar su numero de apariciones y al terminar el bucle, moda contendrá el valor con mayor número de apariciones y lo devuelve como resultado.



```
def moda(arreglo):
    valores_unicos = set(arreglo)
    base = 0

    for valor in valores_unicos:
        apariciones = arreglo.count(valor)

#Calcula la varianza y la desviación estándar del conjunto de datos anterior.

Con los datos anteriores de la array creamos ahora la funcion de la varianza y la desviacion estándar.

- La varianza: se crea una funcion con el nombre varianza el cual va recibir el arreglo con los datos anteriores, de ahi creamos una variable llamada promedio, la cual llama a otra función llamada media, que se asume definida previamente, y que calcula el promedio de todos los valores del arreglo, de ahi retorna la suma de todo el proceso dentro de el.

- x - promedio:

  - para cada elemento x del arreglo:

    - Resta el promedio (x - promedio) → esto da la diferencia o desviación respecto al promedio.

    - Luego eleva esa diferencia al cuadrado (** 2) para eliminar los signos negativos y dar más peso a los valores alejados.

- for x in arreglo:

  - Esto forma una expresión generadora, que recorre todos los elementos x de la lista y calcula (x - promedio) ** 2 para cada uno.

```
def varianza(arreglo):
  promedio = media(arreglo)
  return sum((x - promedio) ** 2 for x in arreglo) / len(arreglo)
```

- La desviacion estandar: se crea la funcion para esta que reciba los datos del arreglo, luego de eso se crea una variable media que usa la función np.mean() de la librería NumPy (np), que calcula la media aritmética de todos los valores del arreglo y guarda en n la cantidad total de elementos que hay en el arreglo, inicializa una variable llamada dist_cuadrada en 0, comienza un bucle que recorre cada valor (dato) dentro del arreglo, dato media calcula la diferencia entre el valor actual y la media.
  - Esto indica cuánto se aleja ese dato del promedio.
  Puede ser positivo o negativo.
  - (dato - media) ** 2 eleva la diferencia al cuadrado.
  - Esto elimina el signo negativo (porque al cuadrar todo da positivo).
  - Da más peso a las diferencias grandes (los valores más alejados cuentan más).
  - dist_cuadradas += ... suma esa diferencia cuadrada al acumulador.
  - Después de recorrer todo el arreglo, dist_cuadradas contendrá la suma total de todas las diferencias cuadradas..

- y ya luego devolveria el resultado final de la desviacion estandar.

```
def desviacion_est(arreglo):
  media = np.mean(arreglo)
  n = len(arreglo)
  dist_cuadradas = 0

  for dato in arreglo:
     dist_cuadradas += (dato - media) ** 2

  return (dist_cuadradas / (n - 1)) ** 0.5
```

````python
import numpy as np

arreglo1 = [2, 4, 4, 6, 8, 8, 8, 10]

def media(arreglo):
  return sum(arreglo) / len(arreglo);

def mediana(arreglo):
  arreglo_ordenado = sorted(arreglo);
  len_arreglo = len(arreglo);

  if(len_arreglo % 2 != 0):
    idx = int(len_arreglo / 2)
    return arreglo_ordenado[idx]
  else:
    idx = int(len_arreglo / 2)
    return (arreglo_ordenado[idx] + arreglo_ordenado[idx - 1]) / 2

def moda(arreglo):
    valores_unicos = set(arreglo)
    base = 0

    for valor in valores_unicos:
        apariciones = arreglo.count(valor)
        if apariciones > base:
            moda = valor
            base = apariciones

    return moda

def desviacion_est(arreglo):
  media = np.mean(arreglo)
  n = len(arreglo)
  dist_cuadradas = 0

  for dato in arreglo:
     dist_cuadradas += (dato - media) ** 2

  return (dist_cuadradas / (n - 1)) ** 0.5

def varianza(arreglo):
  promedio = media(arreglo)
  return sum((x - promedio) ** 2 for x in arreglo) / len(arreglo)

print("\n=== PUNTO 2 ===")
print("La media del arreglo es: ", media(arreglo1))
print("La mediana del arreglo es: ", mediana(arreglo1))
print("La moda del arreglo es: ", moda(arreglo1))
print("\n=== PUNTO 3 ===")
print("La desviacion estandar es de: ", desviacion_est(arreglo1))
print("La varianza es de: ", varianza(arreglo1))
````

PUNTO 9.

El objetivo es dividir el conjunto de datos de estudiantes por género y calcular la media de la columna Nota para cada grupo resultante.

1. Preparación de la Columna de Segmentación (Género)
Aunque el requisito es "agregar" la columna Género, el código demuestra que esta columna ya está presente en el DataFrame cargado desde el archivo Excel (df). Esta columna es crucial, ya que provee la variable categórica necesaria para la segmentación.

2. Aplicación de la Estrategia Split-Apply-Combine
Se emplea la metodología estándar de Pandas para análisis segmentado:

Split (Dividir): El DataFrame df se divide en grupos, donde cada grupo contiene filas que comparten el mismo valor en la columna especificada (Género).
Apply (Aplicar): Dentro de cada grupo (e.g., el grupo 'F' y el grupo 'M'), se selecciona la columna Nota y se aplica la función de agregación mean() (media o promedio).
Combine (Combinar): Pandas combina los resultados de la agregación de cada grupo en una nueva Serie, donde el índice es la columna de agrupación (Género) y los valores son los promedios calculados.

3. Resultados
La Serie resultante muestra el promedio de las notas para cada categoría de género presente en los datos sintéticos. Este resultado indica que, en el conjunto de datos de este taller, los estudiantes clasificados como 'M' tienen una nota promedio ligeramente superior a los estudiantes clasificados como 'F' ($\approx 4.33$ vs. $4.20$).

````python
if "Género" not in df.columns:
    df["Género"] = np.random.choice(["F", "M"], size=len(df))

promedio_genero = df.groupby("Género")["Nota"].mean()

print("\nPUNTO 9")
print("Promedio de notas por género:")
display(promedio_genero)


rango = df["Nota"].max() - df["Nota"].min()
varianza = df["Nota"].var()
desviacion_std = df["Nota"].std()

print("\n=== PUNTO 10 ===")
print(f"Rango: {rango:.2f}")
print(f"Varianza: {varianza:.2f}")
print(f"Desviación estándar: {desviacion_std:.2f}")

plt.figure(figsize=(6,4))
plt.hist(df["Nota"], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución de las Notas")
plt.xlabel("Notas")
plt.ylabel("Frecuencia")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
````
