- La mediana: la funcion de mediana crea un arreglo con la funcion sorted que lo organiza de forma ascendente con los datos del arreglo creado anteriormente y luego se saca la longitud del arreglo, de ahi creamos un condicional que dividir√° la longitud en 2 y si el resto es distinto a 0, su longitud va ser impar, entonces se crea un indice y calculara el indice del elemento central y devuelve el valor de la posicion que marcaba el indice si al dividir es par va a calcular el indice superior del par central va tomar los dos valores centrales, los suma y divide la suma entre 2 y eso produce la mediana si es par

```
def mediana(arreglo):
  arreglo_ordenado = sorted(arreglo);
  len_arreglo = len(arreglo);

  if(len_arreglo % 2 != 0):
    idx = int(len_arreglo / 2)
    return arreglo_ordenado[idx]

    idx = int(len_arreglo / 2)
```

#Calcula la varianza y la desviaci√≥n est√°ndar del conjunto de datos anterior.

Con los datos anteriores de la array creamos ahora la funcion de la varianza y la desviacion est√°ndar.

- La varianza: se crea una funcion con el nombre varianza el cual va recibir el arreglo con los datos anteriores, de ahi creamos una variable llamada promedio, la cual llama a otra funci√≥n llamada media, que se asume definida previamente, y que calcula el promedio de todos los valores del arreglo, de ahi retorna la suma de todo el proceso dentro de el.

- x - promedio:

  - para cada elemento x del arreglo:

    - Resta el promedio (x - promedio) ‚Üí esto da la diferencia o desviaci√≥n respecto al promedio.

    - Luego eleva esa diferencia al cuadrado (** 2) para eliminar los signos negativos y dar m√°s peso a los valores alejados.

- for x in arreglo:

  - Esto forma una expresi√≥n generadora, que recorre todos los elementos x de la lista y calcula (x - promedio) ** 2 para cada uno.

```
def varianza(arreglo):
  promedio = media(arreglo)
  return sum((x - promedio) ** 2 for x in arreglo) / len(arreglo)
```

- La desviacion estandar: se crea la funcion para esta que reciba los datos del arreglo, luego de eso se crea una variable media que usa la funci√≥n np.mean() de la librer√≠a NumPy (np), que calcula la media aritm√©tica de todos los valores del arreglo y guarda en n la cantidad total de elementos que hay en el arreglo, inicializa una variable llamada dist_cuadrada en 0, comienza un bucle que recorre cada valor (dato) dentro del arreglo, dato media calcula la diferencia entre el valor actual y la media.
  - Esto indica cu√°nto se aleja ese dato del promedio.
  Puede ser positivo o negativo.
  - (dato - media) ** 2 eleva la diferencia al cuadrado.
  - Esto elimina el signo negativo (porque al cuadrar todo da positivo).
  - Da m√°s peso a las diferencias grandes (los valores m√°s alejados cuentan m√°s).
  - dist_cuadradas += ... suma esa diferencia cuadrada al acumulador.
  - Despu√©s de recorrer todo el arreglo, dist_cuadradas contendr√° la suma total de todas las diferencias cuadradas..

- y ya luego devolveria el resultado final de la desviacion estandar.

```
def desviacion_est(arreglo):
  media = np.mean(arreglo)
  n = len(arreglo)
  dist_cuadradas = 0

  for dato in arreglo:
     dist_cuadradas += (dato - media) ** 2

  return (dist_cuadradas / (n - 1)) ** 0.5
```

````python
import numpy as np

arreglo1 = [2, 4, 4, 6, 8, 8, 8, 10]

def media(arreglo):
  return sum(arreglo) / len(arreglo);

def mediana(arreglo):
  arreglo_ordenado = sorted(arreglo);
  len_arreglo = len(arreglo);

  if(len_arreglo % 2 != 0):
    idx = int(len_arreglo / 2)
    return arreglo_ordenado[idx]
  else:
    idx = int(len_arreglo / 2)
    return (arreglo_ordenado[idx] + arreglo_ordenado[idx - 1]) / 2

def moda(arreglo):
    valores_unicos = set(arreglo)
    base = 0

    for valor in valores_unicos:
        apariciones = arreglo.count(valor)
        if apariciones > base:
            moda = valor
            base = apariciones

    return moda

def desviacion_est(arreglo):
  media = np.mean(arreglo)
  n = len(arreglo)
  dist_cuadradas = 0

  for dato in arreglo:
     dist_cuadradas += (dato - media) ** 2

  return (dist_cuadradas / (n - 1)) ** 0.5

def varianza(arreglo):
  promedio = media(arreglo)
  return sum((x - promedio) ** 2 for x in arreglo) / len(arreglo)

print("\n=== PUNTO 2 ===")
print("La media del arreglo es: ", media(arreglo1))
print("La mediana del arreglo es: ", mediana(arreglo1))
print("La moda del arreglo es: ", moda(arreglo1))
print("\n=== PUNTO 3 ===")
print("La desviacion estandar es de: ", desviacion_est(arreglo1))
print("La varianza es de: ", varianza(arreglo1))
````

PUNTO 9.

El objetivo es dividir el conjunto de datos de estudiantes por g√©nero y calcular la media de la columna Nota para cada grupo resultante.

1. Preparaci√≥n de la Columna de Segmentaci√≥n (G√©nero)
Aunque el requisito es "agregar" la columna G√©nero, el c√≥digo demuestra que esta columna ya est√° presente en el DataFrame cargado desde el archivo Excel (df). Esta columna es crucial, ya que provee la variable categ√≥rica necesaria para la segmentaci√≥n.

2. Aplicaci√≥n de la Estrategia Split-Apply-Combine
Se emplea la metodolog√≠a est√°ndar de Pandas para an√°lisis segmentado:

Split (Dividir): El DataFrame df se divide en grupos, donde cada grupo contiene filas que comparten el mismo valor en la columna especificada (G√©nero).
Apply (Aplicar): Dentro de cada grupo (e.g., el grupo 'F' y el grupo 'M'), se selecciona la columna Nota y se aplica la funci√≥n de agregaci√≥n mean() (media o promedio).
Combine (Combinar): Pandas combina los resultados de la agregaci√≥n de cada grupo en una nueva Serie, donde el √≠ndice es la columna de agrupaci√≥n (G√©nero) y los valores son los promedios calculados.

3. Resultados
La Serie resultante muestra el promedio de las notas para cada categor√≠a de g√©nero presente en los datos sint√©ticos. Este resultado indica que, en el conjunto de datos de este taller, los estudiantes clasificados como 'M' tienen una nota promedio ligeramente superior a los estudiantes clasificados como 'F' ($\approx 4.33$ vs. $4.20$).

````python
if "G√©nero" not in df.columns:
    df["G√©nero"] = np.random.choice(["F", "M"], size=len(df))

promedio_genero = df.groupby("G√©nero")["Nota"].mean()

print("\nPUNTO 9")
print("Promedio de notas por g√©nero:")
display(promedio_genero)


rango = df["Nota"].max() - df["Nota"].min()
varianza = df["Nota"].var()
desviacion_std = df["Nota"].std()

print("\n=== PUNTO 10 ===")
print(f"Rango: {rango:.2f}")
print(f"Varianza: {varianza:.2f}")
print(f"Desviaci√≥n est√°ndar: {desviacion_std:.2f}")

plt.figure(figsize=(6,4))
plt.hist(df["Nota"], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribuci√≥n de las Notas")
plt.xlabel("Notas")
plt.ylabel("Frecuencia")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
````
